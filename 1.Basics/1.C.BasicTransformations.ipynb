```python
import pandas as pd
```

```python
insurance_pandas_df = pd.read_csv("/dbfs/FileStore/SampleDataSets/insurance.csv")

insurance_df = spark.createDataFrame(insurance_pandas_df)
```

```python
insurance_df.rdd.getNumPartitions()
```

```python
#Show simply displays the first 20 rows in your Dataframe
insurance_df.show()
```

```python
insurance_df.display()
```

```python
insurance_df.limit(20).display()
```

```python
insurance_df.printSchema()
```

```python
#Select oprations can be peformed completely in memory and do not require a shuffle write to disk
insurance_df.select('age', 'sex', 'smoker').display(5)
```

```python
#The describe gives you summary statistics from your DataFrame such as count, mean, standard deviation, min, max
insurance_df.describe().display()
```

```python

insurance_df.select('sex').distinct().show()
```

```python
insurance_df.crosstab('sex', 'smoker')\
            .select('sex_smoker','yes','no')\
            .display()
```

```python
insurance_df.select('age', 'sex', 'bmi', 'region', 'charges')\
            .filter(insurance_df['region'] == 'southwest')\
            .limit(10)\
            .display()
```

```python
insurance_df.select('age', 'sex', 'bmi', 'region', 'charges')\
            .where(insurance_df['age'] > 50)\
            .limit(10)\
            .display()
```

```python
insurance_df.select('age', 'sex', 'bmi', 'region', 'charges')\
            .where((insurance_df['age'] > 50) & (insurance_df['sex'] == 'female'))\
            .orderBy('charges')\
            .display()
```

```python
insurance_df.select('age', 'sex', 'bmi', 'region', 'charges')\
            .where(insurance_df['region'].isin(['southwest', 'southeast']))\
            .withColumnRenamed('sex', 'gender')\
            .orderBy('charges')\
            .display()
```
